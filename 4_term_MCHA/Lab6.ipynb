```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
```

```python
k = 3
m = 1.5
```

```python
X = np.linspace(0, 1, 11)
p = [0, 0.41, 0.79, 1.13, 1.46, 1.76, 2.04, 2.3, 2.55, 2.79, 3.01]
Y = [p_i+(-1)**k*m for p_i in p]
```

```python
X = [0, 1/3, 1]
Y = [0, 1/2, 1]
```

```python
plt.plot(X, Y)
plt.show()
```

```python
eq = 0
for i in range(len(X)):
    temp = 1
    for j in range(len(X)):
        if i != j:
            temp *=(x-X[j])/(X[i]-X[j])
    eq += temp * Y[i]
    print(eq)
eq = sympy.expand(eq)
eq
```

```python
y = [eq.evalf(subs={x: xx}) for xx in X]
```

### Newton

```python
pairs = []
for i in range(len(X)):
    pairs.append((X[i], Y[i]))
```

```python
pairs
```

```python
def get_f(vec):
    if len(vec) == 2:
        return (vec[1][1] - vec[0][1])/(vec[1][0] - vec[0][0])
    return (get_f(vec[1:])-get_f(vec[:-1]))/(vec[-1][0] - vec[0][0])
```

```python
eq = Y[0]
for i in range(1, len(X)):
    temp = 1
    for j in range(i):
        temp *= (x-X[j])
    eq += temp * get_f(pairs[:i+1])
    print(eq)
eq = sympy.expand(eq)
eq
```

```python
y = [eq.evalf(subs={x: xx}) for xx in X]
```

```python
eq.evalf(subs={x: 0.47})
```

```python
plt.figure(figsize=(10,10))
plt.plot(X, y)
plt.plot(X, Y)
plt.legend(["Newton", "func"])
plt.show()
```

```python
def get_f_from_polinom(polinom):
    def f(x):
        val = polinom[0]
        for i in range(len(polinom)-1):
            val = val*x+polinom[i+1]
        return val
    return f
```

```python
def eval_error(X, Y, f):
    er = 0
    for i in range(len(X)):
        er += (Y[i] - f(X[i]))**2
    return er
```

```python
eval_error(X, Y, f)
```

```python
def get_row(x0, n):
    l = []
    for i in range(n + 1):
        l.append(x0**i)
    return np.array(l)
```

```python
for i in range(1, len(Y)+4):
    n0 = i
    F = np.array([get_row(x0, n0) for x0 in X])
    inverse_polinom = np.linalg.lstsq(F, Y)[0]
    print(f" {n0} ,  = {eval_error(X, Y, get_f_from_polinom(inverse_polinom[::-1]))}")
```
